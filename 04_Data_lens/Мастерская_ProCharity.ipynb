```python
  def observe_data(table):
      '''Giving common understanding of data in dataframe'''
      print ('Случайная запись таблицы')
      display(table.sample())
      print('_'*100)
      print(f'Количество записей в таблице: {table.shape[0]} \nKоличество столбцов в таблице: {table.shape[1]}')
      print('_'*100)
      names_columns = table.columns
      print(f'Названия столбцов: {names_columns}')
      print('_'*100)
      names_to_corrige = []
      for name_column in names_columns:
          if not name_column.islower() or ' ' in name_column:
              names_to_corrige.append(name_column)
      if len(names_to_corrige) == 0:
          print ('Название столбцов корректно')
      else:
          print(f'Название следующих колонок должны быть откорретированы: {names_to_corrige}')
      print('_'*100)
      lost_values = table.isna().sum()
      missed_values = lost_values.to_frame(name='missed_values')
      missed_values['percent'] = round(table.isna().mean()*100, 0)
      print ('Пропущенные значения в абсолютном и относительном выражениях:')
      display(missed_values)
      print('_'*100)
      print('Типы данных в колонках:')
      print(table.dtypes)
      print('_'*100)
      print(f'Количество явных дубликатов: {table.duplicated().sum()}', 
            f'что составляет {round(table.duplicated().mean()*100, 0)} % от всех записей')
```

```python
observe_data(volunteers)
```

```python
from datetime import datetime, timedelta
```

```python
volunteers['Date_Created'] =  pd.to_datetime(volunteers['Date_Created'])
```

```python
volunteers['Date_Created'] = pd.to_datetime(volunteers['Date_Created'], format = '%d/%m/%Y %H:%M').dt.date
```

```python
for column in ['Date_Moderation', 'Date_FirstRespond', 'Date_ModerationDenial', 'Date_FirstTask','Date_Created']:
    volunteers[column] =  pd.to_datetime(volunteers[column])
```

Нужно удалить пропуски в числа в колонке ['ID_Volunteer'] 

```python
volunteers['ID_Volunteer'] = volunteers['ID_Volunteer'].replace(r'\s+','',regex=True)
```

```python
volunteers['ID_Volunteer']
```

```python
volunteers['ID_Volunteer'] = volunteers['ID_Volunteer'].astype(str).astype(float)
```

```python
volunteers.dtypes
```

```python
date = datetime.strptime('2020-01-01', '%Y-%m-%d').date()
```

```python
volunteers = volunteers.query('Date_Created >= @date')
```

# Подготовка таблицы для подсчета рейтинга по компаниям

Нам нужно в таблицу рейтинга добавить информацию по названию компании. Для этого из таблицы с волонтерами возьмем ID волонтера  ID компании , сджойним с названием компании и присоединим к тбалице с рейтингом.

```python
temple = volunteers[['ID_Volunteer','ID_Company']]
```

```python
temple.info()
```

```python
temple = temple.merge(company, left_on='ID_Company', right_on='ID_Company')
```

```python
raiting = raiting.merge(temple, left_on='ID_Volunteer', right_on='ID_Volunteer')
```

```python
raiting.info()
```
